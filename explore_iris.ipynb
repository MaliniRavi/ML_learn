```python
import pandas as pd
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

iris_data = datasets.load_iris()
#iris_data
#first : for rows and 2nd is for columns
#iris_data.data[:,:]
#above is same as iris_data since it will print all data
#Now select only 7 rows
iris_data.data[:7,:]
#Now select only 7 rows and 3 column
iris_data.data[:7,:3]
#Now try with different positions
iris_data.data[2:7,1:3]
```

```python
plt.figure(figsize = (10, 7)) 
#extract sepel length
x1 = iris_data.data[:,:0] 
  
plt.hist(x1, bins = 20, color = "green") 
plt.title("Sepal Length in cm") 
plt.xlabel("Sepal_Length_cm") 
plt.ylabel("Count") 
```

```python
plt.figure(figsize = (10, 7)) 
#extract sepel width
x2 = iris_data.data[:,1:2] 
  
plt.hist(x2, bins = 20, color = "green") 
plt.title("Sepal Width in cm") 
plt.xlabel("Sepal_Width_cm") 
plt.ylabel("Count") 
```

```python
plt.figure(figsize = (10, 7)) 
#extract Petal Length
x3 = iris_data.data[:,2:3] 
  
plt.hist(x3, bins = 20, color = "green") 
plt.title("Petal Length in cm") 
plt.xlabel("Petal_Length_cm") 
plt.ylabel("Count") 
```

```python
plt.figure(figsize = (10, 7)) 
#extract Petal Width
x4 = iris_data.data[:,3:4] 
  
plt.hist(x4, bins = 20, color = "green") 
plt.title("Petal Width in cm") 
plt.xlabel("Petal_Width_cm") 
plt.ylabel("Count") 
```

```python
X = iris_data.data[:, 0:2]
y = iris_data.target

#print (X,y)

#plt.figure(figsize=(15, 5))
#plt.subplot(121)
#plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
#plt.xlim(xx.min(), xx.max())
plt.title('Scatter plot between Sepal length and Sepal Width')
```
